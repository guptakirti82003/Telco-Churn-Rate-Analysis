```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['ChurnFlag'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df.info()
```

```python
df.groupby('Contract')['ChurnFlag'].mean().sort_values(ascending=False)*100
```

```python
df.groupby('Churn')['MonthlyCharges'].mean()
```

```python
df.groupby('Churn')['tenure'].mean()
```

```python
df.select_dtypes(include=['int64', 'float64']).columns
```

```python
corr_matrix = df[['SeniorCitizen', 'tenure', 'MonthlyCharges',
                  'TotalCharges', 'ChurnFlag']].corr()

corr_matrix
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
corr_matrix['ChurnFlag'].sort_values(ascending=False)
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.corr()['ChurnFlag'].sort_values(ascending=False).head(10)
```

```python
bins = [0, 12, 24, 48, 72]
labels = ['0-12', '12-24', '24-48', '48+']
df['TenureGroup'] = pd.cut(df['tenure'], bins=bins, labels=labels)
```

```python
df.info()
```

```python
df['ChurnLabel'] = df['Churn']
```

```python
df.to_csv("cleaned_telco_churn.csv", index=False)
```
